```python
from google.colab import files

# Upload the file
uploaded = files.upload()
file_name = next(iter(uploaded))

# Load data into a pandas DataFrame
df = pd.read_excel(file_name)

# Display the first few rows to confirm data loading
df.head()
```

Define Function to Handle Queries:

```python
def get_financial_data(company, year, metric):
    # Filter data for the company and year
    company_data = df[(df['Company'] == company) & (df['Year'] == year)]

    if company_data.empty:
        return f"Sorry, no data found for {company} in {year}."

    # Extract the requested metric
    data = company_data.iloc[0]

    if metric == "total revenue":
        return f"The total revenue for {company} in {year} is {data['Total Revenue']}."
    elif metric == "net income":
        return f"The net income for {company} in {year} is {data['Net Income']}."
    elif metric == "cash flow":
        return f"The cash flow from operations for {company} in {year} is {data['Cash Flow from Operations']}."
    elif metric == "total assets":
        return f"The total assets for {company} in {year} is {data['Total Assets']}."
    else:
        return "Sorry, I don't understand the metric you're asking for."

# Test the function with different queries
print(get_financial_data("Microsoft", 2024, "total revenue"))
print(get_financial_data("Tesla", 2024, "net income"))
print(get_financial_data("Apple", 2024, "cash flow"))
```

Query for Year-over-Year Changes:

```python
def get_year_over_year_change(company, year):
    # Filter data for the company and previous year
    company_data = df[(df['Company'] == company)]
    current_year_data = company_data[company_data['Year'] == year]
    previous_year_data = company_data[company_data['Year'] == (year - 1)]

    if current_year_data.empty or previous_year_data.empty:
        return f"Sorry, no data available for {company} in {year} or {year-1}."

    # Calculate the percentage change in total revenue
    current_revenue = current_year_data['Total Revenue'].values[0]
    previous_revenue = previous_year_data['Total Revenue'].values[0]

    revenue_change = ((current_revenue - previous_revenue) / previous_revenue) * 100
    return f"The total revenue for {company} changed by {revenue_change:.2f}% from {year-1} to {year}."

# Test the year-over-year change function
print(get_year_over_year_change("Microsoft", 2024))
print(get_year_over_year_change("Tesla", 2024))
```

```python
# Example queries
print(get_financial_data("Microsoft", 2024, "total revenue"))
print(get_financial_data("Tesla", 2024, "net income"))
print(get_financial_data("Apple", 2024, "cash flow"))
print(get_year_over_year_change("Microsoft", 2024))
```
